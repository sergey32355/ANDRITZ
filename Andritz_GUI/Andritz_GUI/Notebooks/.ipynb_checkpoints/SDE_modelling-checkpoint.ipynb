```python
PATH="c:\\tmp\\50samp_points_chans_1_2_3_PREPROC_mel_rel.pkl"
#df=pd.read_pickle(PATH)
```

```python
def OpenfeatFromPickle(PATH):
    df=pd.read_pickle(PATH)
    cols=df.columns
    cols = list(cols[3:len(cols)])
    lk=df[cols].to_numpy()[:,None]#.tolist()
    if(len(np.shape(lk))==3):lk = lk[:, 0, :]        
    kj=df[df.columns[0]]
    plate_name=list()
    plate_feat=list()
    
    for l in range(0,len(kj)):
        curp_n=kj[l]
        if(curp_n in (plate_name)):
            plate_feat[-1].append(np.asarray(lk[l]))
        else:
            plate_name.append(curp_n)
            plate_feat.append([])
    print("")
    print("******************************************************")
    print("Opened feratures for file: "+str(PATH))
    print("Data shape: "+str(np.shape(np.asarray(lk))))
    print("Plates num: "+str(len(plate_name)))        
    print("Plates names: "+str(plate_name))
    print("******************************************************")
    print("")
    return plate_name,plate_feat
pl_n,pl_feat = OpenfeatFromPickle(PATH)
```

```python
#https://sde.readthedocs.io/en/latest/sde.html
flat_feat=[]
for k in range(0,len(pl_feat)):
    jk=np.concatenate(pl_feat[k]).ravel()
    flat_feat.append(jk)
    print("Signal "+str(k) + " length - "+str(len(jk)))
```

```python
fig = plt.figure(figsize=(5, 4))
ax = fig.add_subplot(111)
for k in range(0,len(flat_feat)):
    ax.plot(flat_feat[k])
plt.show()
```
